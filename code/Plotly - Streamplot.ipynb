```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import datetime as dt
import matplotlib.colors as colors
import datetime as dt
```

```python
full_data_address = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
full_data_df = pd.read_csv(full_data_address,parse_dates = ['date'])
```

```python
my_data_df = full_data_df.pivot_table(index = "date", columns = "location", values = "new_deaths_smoothed")
```

```python
values = my_data_df[["Africa","Asia","Europe","North America","South America","Oceania"]]
x = values.index
my_colors = dict(zip(["Africa","Asia","Europe","North America","South America","Oceania"],
                     ["black","gold","blue","red","orange","green"]))
```

```python
fig = go.Figure()

for k in values.columns:
    
    fig.add_trace(go.Scatter(
        x=x, y=(-values[k])/2,
        legendgroup = k,
        hoverinfo = "skip",
        hovertext = "",
        name = "",
        fillcolor='rgba(0, 0, 0, 0)',
        line=dict(width = 0),
        stackgroup='one'
    ))

for k in values.columns:
    
    fig.add_trace(go.Scatter(
        x=x, y=values[k],
        hoverinfo='text+y',
        hoveron = 'points+fills',
        legendgroup = k,
        fillcolor = "rgba("+",".join([str(x) for x in (np.array(colors.to_rgb(my_colors[k]))*255)])+",0.6)",
        name = k,
        text = k,
        line = dict(width = 0.5, color = "black"),
        stackgroup='one'
    ))

fig.update_layout(hovermode = "x", title = "<b>Daily reported deaths by global region</b><br>"
                  "7-day rolling average. Data: Our World in Data",
                  legend_traceorder="reversed", legend_font = {"size":14},
                  legend_title = "<b>Global region</b>", legend_title_font = {"size":14})

fig.update_xaxes(range = [x.min(),x.max()+dt.timedelta(15)])

fig.show()
```

```python
fig.write_html("world_streamgraph.html")
```
